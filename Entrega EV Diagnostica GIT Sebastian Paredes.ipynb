```python
import json
from collections import Counter
```

```python
with open('farmers-protest-tweets-2021-03-5.json', 'r', encoding='UTF-8') as dataset:
    data = [json.loads(line) for line in dataset]
```

```python
def search(e):
    return e["retweetCount"]

def more_retweet():
    new_data = data[:]
    new_data.sort(reverse=True, key=search)
    list_retweet = new_data[:10]
    count = 1
    for element in list_retweet:
        print(count, "-", element["content"], "\n")
        count += 1
more_retweet()
```

```python
def user_tweet():
    list_user = []
    new_data = data[:]
    for tweet in new_data:
        list_user.append(tweet["user"]["username"])
    list_user_count = Counter(list_user)
    list_user_most = list_user_count.most_common(10)
    count = 1
    for user in list_user_most:
        print(count, "-", user[0], "\n")
        count += 1
user_tweet()
```

```python
def days_tweet():
    list_days = []
    new_data = data[:]
    for tweet in new_data:
        list_days.append(tweet["date"][:10])#SOLO DIA
    list_days_count = Counter(list_days)
    list_days_most = list_days_count.most_common(10)
    count = 1
    for day in list_days_most:
        print(count, "-", day[0], "\n")
        count += 1
days_tweet()
```

```python
def hashtag_tweet():
    list_hashtag = []
    new_data = data[:]
    for content in new_data:
        list_hashtag_com = content["content"].split()
        for item in list_hashtag_com:
            if "#" in item:
                list_hashtag.append(item)
    list_hashtag_count = Counter(list_hashtag)
    list_hashtag_most = list_hashtag_count.most_common(10)
    count = 1
    for hashtag in list_hashtag_most:
        print(count, "-", hashtag[0], "\n")
        count += 1

hashtag_tweet()
```

```python
if __name__ == '__main__':
    print("Los top 10 tweets m√°s retweeted.:\n")
    more_retweet()
    print("______________________________________\n")
    print("Los top 10 usuarios en funci√≥n a la cantidad de tweets que emitieron:\n")
    user_tweet()
    print("______________________________________\n")
    print("Los top 10 d√≠as donde hay m√°s tweets:\n")
    days_tweet()
    print("______________________________________\n")
    print("Top 10 hashtags m√°s usados:\n")
    hashtag_tweet()
    print("______________________________________\n")
```
